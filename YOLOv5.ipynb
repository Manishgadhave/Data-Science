```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

# Path to the output video file
video_path = "/content/yolov5/runs/detect/exp8/street.mp4"

# Create a VideoCapture object
cap = cv2.VideoCapture(video_path)

# Check if the video file was opened correctly
if not cap.isOpened():
    raise FileNotFoundError(f"Cannot open video file: {video_path}")

# Get the video properties
fps = cap.get(cv2.CAP_PROP_FPS)  # Frames per second
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print(f"Video FPS: {fps}")
print(f"Total number of frames: {frame_count}")

# Define a function to display frames
def display_frame(frame):
    # Convert BGR frame to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Display the frame
    plt.figure(figsize=(12, 7))
    plt.imshow(rgb_frame)
    plt.axis('off')
    plt.show()

# Read and display video frames
try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Display the frame
        display_frame(frame)
        clear_output(wait=True)  # Clear previous frame to update with the new one

finally:
    cap.release()  # Release the video capture object
```
